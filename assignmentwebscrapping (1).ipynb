# 2. Display list of former presidents of India 

```python
page=requests.get('https://presidentofindia.nic.in/former-presidents')
```

```python
page
```

# PAGE CONTENT

```python
soup=BeautifulSoup(page.content)
soup
```

# SCRAPPING FIRST NAME

```python
first_title=soup.find('div',class_="desc-sec")
first_title
```

```python
first_title.text
```

```python
titles=[]

for i in soup.find_all('div',class_="desc-sec"):
    titles.append(i.text)   
titles    
```

# SCRAPPING MULTIPLE TITLES

```python
dfs = []

# Splitting on '\n' and adding to the list
for item in titles:
    data = item.strip().split('\n')
    dfs.append(pd.DataFrame([data], columns=['President', 'Title']))

# Concatenate the list of DataFrames
df = pd.concat(dfs, ignore_index=True)

# Print the DataFrame
print(df)
```
